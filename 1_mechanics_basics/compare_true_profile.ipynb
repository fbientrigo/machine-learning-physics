```python
import numpy as np

def song_f1(v):
    """Function used fron generating data for training on song_f1"""
    v = np.abs(v)
    return (v*(300-v)/1000) * ( 1 + np.sin(v/20)/10 + np.cos(v/40)/10 ) + (v/70)**2


def song_f2(v):
    v = np.abs(v)
    return 1.051359*v-4.24475e-2*v**2 +5.03648e-4*v**3+2.73048e-6 *v**4 -9.34265e-8*v**5 + 6.91675e-10*v**6 - 2.16279e-12*v**7 + 2.50634e-15 * v**8

def song_f3(v):
    v = np.abs(v)
    return (2.898644e-1*v-  8.043560e-3*v**2 + 9.985840e-5 *v**3 - 5.537040e-7 *v**4 + 1.284692e-9*v**5 -8.786800e-13*v**6) * (np.tanh((v-25)/20) + np.tanh((75-v)/20) + 1.5)
```
