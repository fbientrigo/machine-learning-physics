```python
import matplotlib.pyplot as plt
import numpy as np
 
def getPositionEncoding(seq_len, d, n=10000):
    P = np.zeros((seq_len, d))
    for k in range(seq_len):
        for i in np.arange(int(d/2)):
            denominator = np.power(n, 2*i/d)
            P[k, 2*i] = np.sin(k/denominator)
            P[k, 2*i+1] = np.cos(k/denominator)
    return P
 
P = getPositionEncoding(seq_len=4, d=4, n=100)
print(P)
```

```python
P = getPositionEncoding(seq_len=100, d=512, n=10000)
cax = plt.matshow(P)
plt.gcf().colorbar(cax)
plt.show()
```

```python
# Las sinusoidales que la componen
def plotSinusoid(k, d=512, n=10000):
    x = np.arange(0, 100, 1)
    denominator = np.power(n, 2*x/d)
    y = np.sin(k/denominator)
    plt.plot(x, y)
    plt.title('k = ' + str(k))
 
fig = plt.figure(figsize=(15, 4))    
for i in range(4):
    plt.subplot(141 + i)
    plotSinusoid(i*4)
```

el producto punto se mantiene similar, osea k que se encontraban proximos, siguen siendo proximos en este espeacio dimensional latenten

## Obtener PE de una grilla tensorial

```python
from torch import Tensor, meshgrid
```

```python
xL, xR = -5, 5
yL, yR = -5, 5
Npoints = 100

x = Tensor([x for x in np.linspace(xL, xR, Npoints)])
y = Tensor([y for y in np.linspace(yL, yR, Npoints)])

grid_x, grid_y = meshgrid(x,y)


cax = plt.matshow(grid_x + grid_y)
plt.gcf().colorbar(cax)
plt.show()
```
