```python
# vamos a separar el numero de data en factores primos
def primes(n):
    primfac = []
    d = 2
    while d*d <= n:
        while (n % d) == 0:
            primfac.append(d)  # supposing you want multiple factors repeated
            n //= d
        d += 1
    if n > 1:
       primfac.append(n)
    return primfac

primes(n_data)
```

```python
# al reiniciar el entrenamiento es necesario resetear los parametros de training
def reset_weights(m):
  '''
    Resetea los pesos del modelo para evitar weight leakage.
  '''
  for layer in m.children():
   if hasattr(layer, 'reset_parameters'):
    print(f'Reset trainable parameters of layer = {layer}')
    layer.reset_parameters() 


# tambien definiremos un error cuadratico medio
def loss_L2(output, label):
    """ returns MSE """
    return ( (output - label)**2 ).mean()
```

```python
class myNet(nn.Module):
    def __init__(self):
        super().__init__()
        # Linear funciona como input_f, output_f, bias=bool
        # esta es una forma mas elegante de incluir las capas
        self.layers = nn.Sequential(
            nn.Linear(n_features, 50), #A1
            nn.ReLU(),
            nn.Linear(50, 100), 
            nn.ReLU(),
            nn.Linear(100,1), #A3
        )

    def forward(self, x):
        """ realiza el paso a traves de las capas secuenciales """
        return self.layers(x)
```

```python
model = myNet() # instanciamos el modelo    
```
